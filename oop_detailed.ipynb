## 🚀 Class & Object: "Build It Once, Use It Forever"
Let’s say you’re designing a blueprint for a sneaker. You're nott making one sneaker at a time—you're designing a template that lets you make as many pairs as you need.

That’s what a class does—it’s a blueprint for making objects.





Check it:

```python
class Sneaker:
    def __init__(self, brand, size, color):
        self.brand = brand
        self.size = size
        self.color = color

    def show_info(self):
        return f"This is a {self.color} {self.brand} sneaker, size {self.size}."

# Making different sneaker objects
nike = Sneaker("Nike", 10, "Red")
adidas = Sneaker("Adidas", 11, "Blue")

print(nike.show_info())   # "This is a Red Nike sneaker, size 10."
print(adidas.show_info()) # "This is a Blue Adidas sneaker, size 11."
```

🛠️ Translation:

* The class (`Sneaker`) is like Nike’s factory—it can make many sneakers.
* The objects (`nike`, `adidas`) are like actual pairs of shoes made from the blueprint.
* `self.brand`, `self.size`, `self.color` store individual sneaker details.


## OOP Approach: Create a Base DataTransformer Class with Specific Transformations

```python
class DataTransformer:
    def __init__(self, data):
        self.data = data

    def transform(self):
        raise NotImplementedError("This method must be implemented by subclasses")

# Child class for filtering missing data
class FilterMissingValues(DataTransformer):
    def transform(self):
        return [x for x in self.data if x is not None]  # Remove None values

# Child class for normalizing data
class NormalizeData(DataTransformer):
    def transform(self):
        max_value = max(self.data)
        return [x / max_value for x in self.data]  # Normalize values to 0-1 range

# Child class for adding a derived feature
class AddFeature(DataTransformer):
    def transform(self):
        return [(x, x**2) for x in self.data]  # Adds squared value as a new feature

# Example Usage
data = [10, 20, None, 40, 50]

filter_transform = FilterMissingValues(data)
normalized_transform = NormalizeData(filter_transform.transform())
feature_transform = AddFeature(normalized_transform.transform())

print(feature_transform.transform())
# Output: [(0.2, 0.04), (0.4, 0.16), (0.8, 0.64), (1.0, 1.0)]
```

### OOP Approach: Create a Pipeline Class

```python
class DataPipeline:
    def __init__(self, data_source):
        self.data_source = data_source  # Store the data source
        self.steps = []  # Store processing steps

    def add_step(self, step):
        self.steps.append(step)  # Add a new transformation step

    def run(self):
        data = self.data_source.read_data()  # Read the raw data
        for step in self.steps:
            data = step.transform(data)  # Apply each transformation
        return data  # Return final processed data

# Example usage
csv_source = CSVSource("data.csv")  # Load from CSV

pipeline = DataPipeline(csv_source)
pipeline.add_step(FilterMissingValues)
pipeline.add_step(NormalizeData)
pipeline.add_step(AddFeature)

final_data = pipeline.run()
print(final_data)
```

1️⃣ What is `__init__` and Why Do We Need It?

🔹 Question: What’s the purpose of `__init__` in Python classes?

* `__init__` is a constructor function that runs automatically when an object is created.
* It initializes the object's attributes (stores important values inside the object for later use).
* Every time you create an instance of a class, `__init__` is the first function that gets executed.


```python
class DataTransformer:
    def __init__(self, data):
        self.data = data  # Store data in the object

# Create an instance
transformer = DataTransformer([1, 2, 3, 4, 5])

# Access the data inside the object
print(transformer.data)  # Output: [1, 2, 3, 4, 5]
```

🔹 What’s happening here?

* When you write `transformer = DataTransformer([1, 2, 3, 4, 5])`, Python automatically calls `__init__` and stores the `[1, 2, 3, 4, 5]` list inside `self.data`.
* Now, `transformer.data` lets you access that stored data later.

### 2️⃣ Why Do Some Functions Have `self`, While Others Have Class Names Like `DataTransformer`?
🔹 `self` in Methods
* In Python, functions inside a class must have `self` as the first argument if they modify or access object attributes.
* `self` represents the current instance of the class (it’s like saying "me").
* When you call a method, Python **automatically passes the object itself** as the first argument.


```python
class Example:
    def say_hello(self):
        return "Hello, world!"

obj = Example()
print(obj.say_hello())
```

* Here, `self` refers to `obj`, and calling `obj.say_hello()` runs `say_hello(self)` behind the scenes.

### 🔹 Why Some Classes Use DataTransformer Instead of self
* When defining a new class that inherits from another, you write:

`class FilterMissingValues(DataTransformer):`
This tells Python:
"**This class is a specialized version of DataTransformer.**"

* `self` is used inside a function when referring to an instance, while `DataTransformer` is used when defining a relationship between classes.

### 3️⃣ Why Do Some Classes Inherit from Another (`class FilterMissingValues(DataTransformer`)) While Others Don't?
#### 🔹 Inheritance (When You Want to Reuse and Extend)
* If you have common functionality shared by multiple classes, use inheritance.
* The parent `class (DataTransformer)` contains a general structure.
* The child classes (`FilterMissingValues`, `NormalizeData`) modify or extend this without rewriting everything.

**Example: Using a Parent Class (DataTransformer)**

```python
class DataTransformer:
    def __init__(self, data):
        self.data = data

    def transform(self):
        raise NotImplementedError("This method must be implemented by subclasses")

class FilterMissingValues(DataTransformer):
    def transform(self):
        return [x for x in self.data if x is not None]

class NormalizeData(DataTransformer):
    def transform(self):
        max_value = max(self.data)
        return [x / max_value for x in self.data]

data = [10, 20, None, 40, 50]
filter_step = FilterMissingValues(data)
print(filter_step.transform())  # Output: [10, 20, 40, 50]

normalize_step = NormalizeData(filter_step.transform())
print(normalize_step.transform())  # Output: [0.2, 0.4, 0.8, 1.0]
```

#### Problem: Create a Log Processor System
Imagine you are building a system to process log files from different sources. You have two types of logs:

- System Logs (e.g., syslog.log)
- Application Logs (e.g., app.log)

Your task is to:

- Create a parent class LogProcessor that:

* Stores the filename and log type.
* Defines a method process_log() that must be implemented by child classes.

**Create two child classes:**

- SystemLogProcessor: Should return "Processing system log: {filename}"
- AppLogProcessor: Should return "Processing application log: {filename}"
- Create instances of both SystemLogProcessor and AppLogProcessor, and call process_log() on them.



## My Answer

```python
class LogProcessor:
    def __init__(self, filename, log_type):
        self.filename = filename
        self.log_type = log_type

    def process_log(self):
        raise NotImplementedError("This method must be implemented by subclasses")

class SystemLogProcessor(LogProcessor):
    def process_log(self):
        return f"Processing system {self.log_type}: {self.filename}"

class AppLogProcessor(LogProcessor):
    def process_log(self):
        return f"Processing application {self.log_type}: {self.filename}"

system_log_file = SystemLogProcessor("systemfile", "systemlog")
print(system_log_file.process_log())
app_log_file = AppLogProcessor("appfile", "applog")
print(app_log_file.process_log())
```

OK, so my code does indeed give an answer, but it can be greatly improved.

### Conceptual Mistake: The log_type is Redundant
Problem

In my child classes, I'm explicitly stating "`Processing system {self.log_type}`" and "`Processing application {self.log_type}`".

However, I'm already defining the log type in the class itself. There's no need to pass "`systemlog`" or "`applog`" manually when creating an instance. Instead, I can hardcode the **log type inside each child class** to avoid redundancy.

### Suggested Improvement
Fix: Set `log_type` Inside Each Child Class
Instead of requiring the user to pass `"systemlog"` or `"applog"` manually, I can define it inside the child class and only require `filename` as an argument.

```python
class LogProcessor:
    def __init__(self, filename):
        self.filename = filename  # Store filename

    def process_log(self):
        raise NotImplementedError("This method must be implemented by subclasses")

# ✅ Child classes should set log_type automatically
class SystemLogProcessor(LogProcessor):
    def __init__(self, filename):
        super().__init__(filename)  # Inherit from LogProcessor
        self.log_type = "System Log"  # Set log type automatically

    def process_log(self):
        return f"Processing {self.log_type}: {self.filename}"

class AppLogProcessor(LogProcessor):
    def __init__(self, filename):
        super().__init__(filename)
        self.log_type = "Application Log"

    def process_log(self):
        return f"Processing {self.log_type}: {self.filename}"

# ✅ Create instances
system_log = SystemLogProcessor("syslog.log")
app_log = AppLogProcessor("applog.log")

# ✅ Expected Output
print(system_log.process_log())  # Processing System Log: syslog.log
print(app_log.process_log())     # Processing Application Log: applog.log
```

## Why is This an Improvement?

| **Issue** | **Original Code** | **Improved Code** |
|------------|-----------------|----------------|
| **Redundant argument (`log_type`)** | User has to **manually specify** `"systemlog"` or `"applog"` when creating an instance. | **Log type is set inside the child class**, reducing input errors and redundant code. |
| **Better Encapsulation** | The child classes still depend on external inputs for their type. | Each child class **fully defines itself**, making it self-contained. |
| **Code Clarity** | `"Processing system systemlog: systemfile"` sounds redundant. | `"Processing System Log: syslog.log"` is clearer and structured better. |

### Concept Reinforcement: Understanding `super()`
In our child classes, we wrote:
`super().__init__(filename)`

What does this do?

* `super()` calls the constructor of the parent class (`LogProcessor`).
* This ensures the `filename` is properly initialized in `LogProcessor` before adding additional attributes (`log_type`).

**Key Takeaway:**
Always use `super().__init__()` when your child class needs to initialize attributes from the parent class.
